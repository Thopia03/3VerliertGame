# 2) Problemzerlegung / Algorithmen

Bevor wir zur Modellierung unseres Programmes kommen, müssen wir erst einmal etwas über das Herzstück des Programmes nachdenken: Den Algorithmus zur Bestimmung von 3 nebeneinander stehenden Zeichen.

Dafür müssen wir uns erst einmal konzeptionell in die Problematik eindenken.

- Gegeben ist ein 4x4 Feld, auf welchem abwechselnd Kreuze und Kreise gesetzt werden.
- Verloren hat, wer als erstes drei nebeneinander stehende Zeichen hat, horizontal, vertikal oder diagonal.

Berechtigterweise stellt sich nun ein Problem heraus: während der Case für einen Verlust auf 3x3 basiert, ist das Spielfeld 4x4 groß. Was tun wir also, damit wir vernünftig dies berechnen können?<br><br>
Unser Algorithmus soll folgendes machen:

=> "Finde auf dem 4x4 Spielfeld jede Folge von 3 direkt aufeinanderfolgenden Feldern welche horizontal, vertikal oder diagonal sein können. Dazu definieren wir uns gültige Startblöcke welche für die Überprüfung in Frage kommen, um mit wenig Problemen diese Folgen zu finden."

Um nun die einzelnen Fälle abzudecken und auf Korrektheit zu prüfen, brauchen wir ein mathematisches Konstrukt auf dem wir unseren Algorithmus basieren können.

## <a style = "color: #9B5DE5"> Mathematische Basis: </a>

Konzeptionell definieren wir jetzt gültige "Startpositionen" in unserem Feld, in welchen es Sinn macht zu überprüfen ob wir 3 aufeinanderfolgende Felder haben. Wenn wir zeigen können, dass wir alle Dreierblöcke auf alle gültigen Startpositionen abbilden können und ebenfalls anders herum, dass wir aus allen Startpositionen alle möglichen Dreierblöcke erreichen können, so können wir Garantieren, dass wir mit dieser Methode alle Fälle abdecken. Der Algorithmus wäre dann mathematisch vollständig und korrekt (Diese Abbildung / Funktion zwischen den Mengen muss bijektiv sein).


Definieren wir also folgendes:
- Das Feld besteht aus Koordinaten, welche das Feld aufspannen
- Die Feldkoordinaten bestehen aus dem kartesischen Produkt von X & Y, wobei X = {0, 1, 2, 3} & Y = {0, 1, 2, 3} <br>
<img src = "assets/diagrams/fieldCoordinates.svg"/>
- Sei |X| = 4 als Breite definiert; |Y| = 4 als Höhe definiert
- Wir betrachten die Richtungen horizontal, vertikal, diagonal (nach rechts unten), diagonal (nach links unten)
- horizontal verläuft von links nach rechts, vertikal verläuft von oben nach unten; diagonal (nach rechts unten) verläuft von oben links nach unten rechts & diagonal (nach links unten) verläuft von oben rechts nach unten links <br>
<img src = "assets/diagrams/directionField.svg"/>
- Um unsere Fälle zu betrachten müssen wir uns nun alle möglichen Startpositionen betrachten. Ein "Dreierblock" kann also nur da starten, wo noch zwei weitere Felder in der Richtung existieren. Formal heißt das für alle Fälle:<br><br>
=> Menge horizontale Startblöcke: $$ S_h: \{(x,y) \in X \times Y \mid (x,y), (x+1,y), (x+2,y) \in X \times Y\} = \\ \{(0,0), (1,0), (0,1), (1,1), (0,2), (1,2), (0,3), (1,3)\} $$ <br>
=> Menge vertikale Startblöcke: $$ S_v: \{(x,y) \in X \times Y \mid (x,y), (x,y+1), (x,y+2) \in X \times Y\} = \\ \{(0,0), (0,1), (1,0), (1,1), (2,0), (2,1), (3,0), (3,1)\}$$ <br>
=> Menge diagonale (nach rechts unten) Startblöcke: $$ S_{dru}: \{(x,y) \in X \times Y \mid (x,y), (x+1,y+1), (x+2,y+2) \in X \times Y\} = \\ \{(0,0), (1,0), (0,1), (1,1)\}$$ <br>
=>Menge diagonale (nach links unten) Startblöcke: $$ S_{dlu}: \{(x,y) \in X \times Y \mid (x,y), (x-1,y+1), (x-2,y+2) \in X \times Y\} = \\ \{(2,0), (3,0), (2,1), (3,1)\}$$ <br>

<a style = "color: #2d839b"> (Editors note: für eine zukünftige Änderung für den NxN Fall müssten wir die Definition passend umbauen. Den Vollständigkeitsbeweis, dass für alle möglichen N die Startposition nur dann existieren kann, wenn zwei nachfolgende Felder existieren, könnte z.B.über vollständige Induktion bewiesen werden.

Ebenfalls wäre es theoretisch möglich einen Richtungsvektor einzubauen, mit welchem wir dann einfach später mit einer Formel die jeweiligen möglichen Richtungen gehen könnten.) </a>


Um nun die Korrektheit zu folgern, müssen wir zeigen, dass jede mögliche Dreierfolge genau so eine Startposition besitzt. Nur so ist unser Algorithmus auch vollständig. Da unser Feld recht klein ist, Stelle ich jede mögliche Dreierfolge einem Startblock über. Wenn alle zugeteilt werden können, ist unsere Regel auch korrekt und wir können diese Implementieren als Fallunterscheidungen.

<a style = "color: #2d839b"> (Editors note: Für den NxN reicht diese Definition / der Beweis natürlich nicht aus. Am besten sollten wir hierfür dann eine Funktion definieren für jeden Fall (horizontal, vertikal, etc.), welcher von der Startposition auf das Tripel abbildet. Dann müssten wir noch beweisen, dass die Menge aller Bildmengen von unseren definierten Funktionen und die Menge aller möglichen Dreierfolgen beide gleich sind.) </a>

## <a style = "color: #9B5DE5"> Der Algorithmus logisch beschrieben: </a>

Der später implementierte Algorithmus wird dann so funktionieren:

Der Rückgabewert der Funktion ist ein boolean, true für "Block gefunden" und false für "kein Block vorhanden".

Als Basis für unsere Datenstruktur haben wir ein Grid [x][y] in welchem wir unsere Koordinaten stehen haben. Dieses Grid wird so implementert, dass es 3 mögliche Einträge / Zustände für ein einzelnes Feld in diesem geben kann: <br>
1) "0" = Das Feld ist leer und kein Zeichen ist gesetzt
2) "1" = Spieler A hat das zum Spieler zugehörige Zeichen A gesetzt (Kreuz oder Kreis)
3) "2" = Spieler B hat das zum Spieler zugehörige Zeichen B gesetzt (das jeweils andere was Spieler A nicht besitzt)

Außerdem definieren wir eine Hilfsfunktion, welche am Anfang des Spiels unsere Startpositionen berechnet und in ein Array speichert. Zwar ist dies für den 4x4 Fall nicht zwingend notwendig (sie könnten auch einfach hart-codiert werden), aber interessant für eine mögliche NxN Erweiterung, welche dann dynamisch diese berechnen kann. <br>

Warum ist gerade so eine Hilfsfunktion nun interessant oder gar notwendig?<br>

Wir haben bereits bewiesen, dass unser Konstrukt aus Startpositionen mathematisch vollständig und korrekt ist. Diese Eigenschaft machen wir uns nun zu nutzen, da wir wissen, dass ein Dreierblock nicht ohne Startposition existieren kann. D.h. unser Algorithmus muss erst dann auf einen möglichen Verlust prüfen, wenn ein Zeichen auf eine Startposition gesetzt wurde. <br>

Dies ist effizienter, da der Algorithmus nicht auf allen möglichen Koordinaten ausgefürt wird. Ebenfalls erleichtert dies uns in weniger "Out-of-Bounds"-Fälle potentiell zu kommen, da wir wissen, dass die Fälle horizontal & vertikal nicht über die Grid-Koordinaten hinaus gehen. (Diagonal ist hier ein Spezialfall dann) <br> <br> <br>

Der Algorithmus wird später nach Abschluss des Zuges, des gerade aktiven Spielers, ausgeführt (Siehe Sequenzdiagramm zu Spielzug). Der Algorithmus berechnet dann folgendes:

1) Sammel alle Koordinaten von Einträgen im Grid des aktiven Spielers als Array Ka
2) Filter die Einträge aus Ka auf Startpositionen, indem wir den Schnitt der beiden Mengen bestimmen
3) Wenn der Schnitt leer ist, hören wir mit dem Algorithmus auf (false) => weiter im Sequenzdiagramm
4) Ist der Schnitt nicht leer, so müssen wir uns die Nachbarfelder der Koordinate angucken und schauen, ob dort ebenfalls ein Zeichen gesetzt ist (Wir geben die Koordinate dann an unsere Hilfsfunktionen, welche überprüfen ob ein horizontaler, vertikaler oder diagonaler Block besteht)
5) Ist die Analyse unserer Hilfsfunktionen negativ, so hören wir mit dem Algorithmus auf (false) => weiter im Sequenzdiagramm
6) Ist die Analyse unserer Hilfsfunktion positiv, so geben wir diese Info weiter (true) => weiter im Sequenzdiagramm
