# 2) Problemzerlegung / Algorithmen

Bevor wir zur Modellierung unseres Programmes kommen, müssen wir erst einmal etwas über das Herzstück des Programmes nachdenken: den Algorithmus zur Bestimmung von 3 nebeneinander stehenden Zeichen.

Dafür müssen wir uns erst einmal konzeptionell in die Problematik eindenken.

- Gegeben ist ein 4x4 Feld, auf welchem abwechselnd Kreuze und Kreise gesetzt werden.
- Verloren hat, wer als Erstes drei nebeneinander stehende Zeichen hat, horizontal, vertikal oder diagonal.

Berechtigterweise stellt sich nun ein Problem heraus: während der Case für einen Verlust auf 3x3 basiert, ist das Spielfeld 4x4 groß. Was tun wir also, damit wir vernünftig dies berechnen können?<br><br>
Unser Algorithmus soll folgendes machen:

=> "Finde auf dem 4x4 Spielfeld jede Folge von 3 direkt aufeinanderfolgenden Feldern welche horizontal, vertikal oder diagonal sein können. Dazu definieren wir uns gültige Startblöcke, welche für die Überprüfung infrage kommen, um mit wenig Problemen diese Folgen zu finden."

Um jetzt die einzelnen Fälle abzudecken und auf Korrektheit zu prüfen, brauchen wir ein mathematisches Konstrukt, auf dem wir unseren Algorithmus basieren können.

## <a style = "color: #9B5DE5"> Mathematische Basis: </a>

Konzeptionell definieren wir jetzt gültige "Startpositionen" in unserem Feld, in welchen es Sinn macht, zu überprüfen, ob wir 3 aufeinanderfolgende Felder haben. Wenn wir zeigen können, dass wir alle Dreierblöcke auf alle gültigen Startpositionen abbilden können und ebenfalls andersherum, dass wir aus allen Startpositionen alle möglichen Dreierblöcke erreichen können insgesamt, so können wir garantieren, dass wir mit dieser Methode alle Fälle abdecken. Der Algorithmus wäre dann mathematisch vollständig und korrekt.


Definieren wir also Folgendes:
- Das Feld besteht aus Koordinaten, welche das Feld aufspannen
- Die Feldkoordinaten bestehen aus dem kartesischen Produkt von X & Y, wobei X = {0, 1, 2, 3} & Y = {0, 1, 2, 3} <br>
<img src = "assets/diagrams/fieldCoordinates.svg"/>
- Sei |X| = 4 als Breite definiert; |Y| = 4 als Höhe definiert
- Wir betrachten die Richtungen horizontal, vertikal, diagonal (nach rechts unten), diagonal (nach links unten)
- horizontal verläuft von links nach rechts, vertikal verläuft von oben nach unten; diagonal (nach rechts unten) verläuft von oben links nach unten rechts & diagonal (nach links unten) verläuft von oben rechts nach unten links <br>
<img src = "assets/diagrams/directionField.svg"/>
- Um unsere Fälle zu betrachten, müssen wir uns nun alle möglichen Startpositionen betrachten. Ein "Dreierblock" kann also nur da starten, wo noch zwei weitere Felder in der Richtung existieren. Formal heißt das für alle Fälle:<br><br>
=> Menge horizontaler Startblöcke: $$ S_h: \{(x,y) \in X \times Y \mid (x,y), (x+1,y), (x+2,y) \in X \times Y\} = \\ \{(0,0), (1,0), (0,1), (1,1), (0,2), (1,2), (0,3), (1,3)\} $$ <br>
=> Menge vertikaler Startblöcke: $$ S_v: \{(x,y) \in X \times Y \mid (x,y), (x,y+1), (x,y+2) \in X \times Y\} = \\ \{(0,0), (0,1), (1,0), (1,1), (2,0), (2,1), (3,0), (3,1)\}$$ <br>
=> Menge diagonaler (nach rechts unten) Startblöcke: $$ S_{dru}: \{(x,y) \in X \times Y \mid (x,y), (x+1,y+1), (x+2,y+2) \in X \times Y\} = \\ \{(0,0), (1,0), (0,1), (1,1)\}$$ <br>
=>Menge diagonaler (nach links unten) Startblöcke: $$ S_{dlu}: \{(x,y) \in X \times Y \mid (x,y), (x-1,y+1), (x-2,y+2) \in X \times Y\} = \\ \{(2,0), (3,0), (2,1), (3,1)\}$$ <br>

<a style = "color: #2d839b"> (Editors note: Für eine zukünftige Änderung für den NxN-Fall müssten wir die Definition passend umbauen. Den Vollständigkeitsbeweis, dass für alle möglichen N die Startposition nur dann existieren kann, wenn zwei nachfolgende Felder existieren, könnte z.B.über vollständige Induktion bewiesen werden.

Ebenfalls wäre es theoretisch möglich, einen Richtungsvektor einzubauen, mit welchem wir dann einfach später mit einer Formel die jeweiligen möglichen Richtungen gehen könnten.) </a>


Um nun die Korrektheit zu folgern, müssen wir zeigen, dass jede mögliche Dreierfolge genau so eine Startposition besitzt. Nur so ist unser Algorithmus auch vollständig. Da unser Feld recht klein ist, stelle ich jede mögliche Dreierfolge einem Startblock über. Wenn alle zugeteilt werden können, ist unsere Regel auch korrekt und wir können diese implementieren als Fallunterscheidungen.

<a style = "color: #2d839b"> (Editors note: Für den NxN reicht diese Definition / der Beweis natürlich nicht aus. Am besten sollten wir hierfür dann eine Funktion definieren für jeden Fall (horizontal, vertikal, etc.), welche von der Startposition auf das Tripel abbildet. Dann müssten wir noch beweisen, dass die Menge aller Bildmengen von unseren definierten Funktionen und die Menge aller möglichen Dreierfolgen beide gleich sind.) </a>

((1,0), (2,0), (3,0)) -> (1,0) <br>

## <a style = "color: #9B5DE5"> Der Algorithmus logisch beschrieben: </a>

Der später implementierte Algorithmus wird dann so funktionieren:

Der Rückgabewert der Funktion ist ein Boolean, true für "Block gefunden" und false für "kein Block vorhanden".

Als Basis für unsere Datenstruktur haben wir ein Grid [x][y] in welchem wir unsere Koordinaten stehen haben. Dieses Grid wird so implementiert, dass es 3 mögliche Einträge / Zustände für ein einzelnes Feld in diesem geben kann: <br>
1) "0" = Das Feld ist leer und kein Zeichen ist gesetzt
2) "1" = Spieler A hat das zum Spieler zugehörige Zeichen A gesetzt (Kreuz oder Kreis)
3) "2" = Spieler B hat das zum Spieler gehörige Zeichen B gesetzt (das jeweils andere, was Spieler A nicht besitzt)

Außerdem definieren wir eine Hilfsfunktion, welche am Anfang des Spiels unsere Startpositionen berechnet und in einem Array speichert. Zwar ist dies für den 4x4-Fall nicht zwingend notwendig (sie könnten auch einfach hart codiert werden), aber interessant für eine mögliche NxN-Erweiterung, welche dann dynamisch diese berechnen kann. <br>

Warum ist gerade so eine Hilfsfunktion nun interessant oder gar notwendig?<br>

Wir haben bereits bewiesen, dass unser Konstrukt aus Startpositionen mathematisch vollständig und korrekt ist. Diese Eigenschaft machen wir uns nun zunutze, da wir wissen, dass ein Dreierblock nicht ohne Startposition existieren kann. D.h. unser Algorithmus muss erst dann auf einen möglichen Verlust prüfen, wenn ein Zeichen auf eine Startposition gesetzt wurde. <br>

Dies ist effizienter, da der Algorithmus nicht auf allen möglichen Koordinaten ausgeführt wird. Ebenfalls erleichtert dies uns, in weniger "Out-of-Bounds"-Fälle potentiell zu kommen, da wir wissen, dass die Fälle horizontal & vertikal nicht über die Grid-Koordinaten hinausgehen (Diagonal ist hier ein Spezialfall dann). <br> <br> <br>

Der Algorithmus wird später nach Abschluss des Zuges des gerade aktiven Spielers ausgeführt (Siehe Sequenzdiagramm zu Spielzug). Der Algorithmus berechnet dann Folgendes:

1) Sammle alle Koordinaten von Einträgen im Grid des aktiven Spielers als Array Ka
2) Filter die Einträge aus Ka auf Startpositionen, indem wir den Schnitt der beiden Mengen bestimmen
3) Wenn der Schnitt leer ist, hören wir mit dem Algorithmus auf (false) => weiter im Sequenzdiagramm
4) Ist der Schnitt nicht leer, so müssen wir uns die Nachbarfelder der Koordinate angucken und schauen, ob dort ebenfalls ein Zeichen gesetzt ist (wir geben die Koordinate dann an unsere Hilfsfunktionen, welche überprüfen, ob ein horizontaler, vertikaler oder diagonaler Block besteht)
5) Ist die Analyse unserer Hilfsfunktionen negativ, so hören wir mit dem Algorithmus auf (false) => weiter im Sequenzdiagramm
6) Ist die Analyse unserer Hilfsfunktion positiv, so geben wir diese Info weiter (true) => weiter im Sequenzdiagramm

## <a style = "color: #9B5DE5"> Laufzeit </a>

Ebenfalls stellt sich die Frage, ob der Algorithmus mit den 'validen Startpositionen' wirklich für einen NxN-Fall effizienter wäre.

Wir schauen uns den Algorithmus einmal an:<br><br><br>

<a style = "color: #83b818"> "1) Collect all coordinates of the symbols placed by the current player.": </a><br><br>
Im Worst-Case bauen wir uns ein Feld mit der Größe $O\!\left(\frac{n^2}{2}\right) = O(n^2)$ <br>
Ebenfalls gehen wir durch das ganze Feld, dementsprechend haben wir hier auf jeden Fall eine Laufzeit von $O(n^2)$

<a style = "color: #83b818"> "2) Now we filter the list by intersecting the coordinates with the valid start positions.": </a><br><br>
Wir definieren 
- A = playerCoordinates
- B = validStartPositions

Die Komplexität unseres Schnittes ist $O(\min(\lvert A\rvert, \lvert B\rvert))$
Für ein NxN-Feld gilt, dass validStartPositions $\approx O(n^2)$, da wir nur konstant 25 % einsparen und die Anzahl der möglichen Startpunkte quadratisch wächst.<br>
Dementsprechend gilt für Worst-Case an Speicherplatz: $O(\min(\lvert A\rvert, \lvert B\rvert)) = O(\min(O(n^2), O(n^2))) = O(n^2)$ (auch für Laufzeit, da wir quadratisch große Mengen miteinander vergleichen).

3 brauchen wir nicht zu analysieren, da es eine einfache Überprüfung ist.

<a style = "color: #83b818"> "4) Now that there are valid start positions, we need to check their neighbours for a hit." </a><br><br>
Wir haben hier vier Methoden, welche die Überprüfung durchführen auf unseren playerCoordinates:
- checkHorizontal
- checkVertical
- checkDiagonalRightDown
- CheckDiagonalLeftDown

Worst-case haben wir hier $\lvert playerCoodinates\rvert = O(n^2)$ viele Koordinaten die wir überprüfen in der Laufzeit.
Da die Funktionen selber nur konstant viele Feldzugriffe (zwei) machen, bleibt die Laufzeit $O(n^2)$.
Auch die Anzahl der Überprüfungen (vier) bleibt gleich, dementsprechend bleibt unsere Laufzeit hier $O(n^2)$

Da die anderen beiden Cases wie 3 einfache Überprüfungen für "checkField" sind, können wir die Gesamtlaufzeit berechnen.

<a style = "color: #83b818"> Gesamtlaufzeit </a><br><br>
Wenn wir alle Cases zusammenführen, berechnet sich unsere Laufzeit als: $T(n) = O(n^2) + O(n^2) + O(n^2) = O(n^2)$

Unser Speicherverbrauch bezieht sich durch die beiden Strukturen "playerCoordinates" und "ValidStartPositions" auf $S(n) = O(n^2) + O(n^2) = O(n^2)$

Für unseren 4x4-Fall ist dies unproblematisch, da wir konstant hier bleiben.


Ist unser Algorithmus nun unnötig kompliziert?

Meiner Meinung nach ist der Algorithmus ein interessanter Approach bezüglich des Problems, er spart jedoch nichts an Komplexität ein.
Alternativ könnte man auch einfach einen einfachen Algorithmus beschreiben, welcher nur vom zuletzt plazierten Feld aus prüft.
Dies wäre für einen NxN-Fall wahrscheinlich einfacher zu lösen und weniger komplex, allerdings würde auch hierzu eine vernünftige Analyse fällig.
